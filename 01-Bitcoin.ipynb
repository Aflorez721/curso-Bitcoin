## 1. ¿Qué es Bitcoin?

> "Bitcoin is a completely new, unprecedented, and complex technology" (Antonopoulos, A. 2015. pp: 236).

En terminos simples, Bitcoin es una técnologia pero también es una moneda. Pero es una moneda muy particular, ya que sólo existe en el mundo digital, por lo que también es un nuevo tipo de moneda llamada **"moneda virtual"** y además posee una caracteristica que la convierte en disruptiva, es **descentralizada**.

Una definición más formal sobre Bitcoin es:

> El bitcoin es una tecnologı́a peer-to-peer (P2P) que permite realizar transferencias online, criptográficamente seguras, de forma directa y sin la intervención de ningún intermediario. La unidad de transferencia de esta plataforma también se denomina bitcoin y se considera una moneda virtual, una moneda digital o una cripto-moneda. 

***
**Observación:**

El termino *Bitcoin* se refiere solo al sistema monetario entero, mientras que *bitcoins* son las unidades basicas de esta moneda (Barski y Wilmer, 2015).

### 1.1 Genesis del Bitcoin


La historia de Bitcoin nace en noviembre de 2008, de la mano de Satoshi Nakamoto en su artículo titulado *Bitcoin: Un sistema de efectivo electrónico de Peer-to-Peer* públicado en una lista de distribución de criptografía. En este documento se detalla cómo usar una red peer-to-peer (P2P) para crear *"un sistema para transacciones electrónicas que no dependa de la confianza"*. En enero de 2009, Satoshi Nakamoto registra el primer bloque (conocido como el bloque de genesis), creando así la red Bitcoin y la emisión de los primeros bitcoins, así como el lanzamiento del primer cliente Bitcoin de código abierto. 

Por su parte, el programador Hal Finney fue una de las primeras personas en adoptar, apoyar y contribuir a Bitcoin. Finney descargó el software Bitcoin el día en que fue lanzado, y recibió 10 bitcoins de Nakamoto en la primera transacción bitcoin del mundo. Otros de los primeros fueron Wei Dai, creador de b-money, y Nick Szabo, creador de bit gold, ambos predecesores de Bitcoin.

En los primeros días, se estima que Nakamoto minó un millón de bitcoins. Antes de desaparecer, Nakamoto en cierto sentido entregó las riendas al desarrollador Gavin Andresen, quien luego se convirtió en el desarrollador líder de Bitcoin en la Fundación Bitcoin, la comunidad "anárquica" de Bitcoin más cercana a una cara pública oficial.

Las primeras transacciones en bitcoins se negociaron entre individuos en los foros de bitcointalk con una notable transacción de $10.000 BTC$ usados para comprar indirectamente dos pizzas entregadas por Papa John's.


### 1.2  Su creador

> Nunca se le ha puesto cara ni voz y ni siquiera se sabe si su nombre es real o inventado.

**Satoshi Nakamoto**, tras escribir el artículo que impulsó el Bitcoin, y se ha convertido en un mito del siglo XXI. Su identidad ha sido objeto de especulación en todo el mundo, y desde los simples aficionados hasta los grandes medios y personalidades diversas han tenido alguna suposición al respecto.

En su perfil de la Fundación P2P, Nakamoto afirmó ser un varón de 37 años de edad residente en Japón, pero algunos especularon que era poco probable que su origen fuera japonés debido a su perfecto uso del inglés y por el hecho de que el software Bitcoin no está ni documentado ni etiquetado en japonés.

Para el año 2010, Satoshi Nakamoto se desvincula del proyecto Bitcoin. Entrega el control del repositorio del código fuente y la clave de alerta de red a Gavin Andresen, desarrollador de Bitcoin, transfiere dominios relacionados a otros miembros importantes de la comunidad Bitcoin y deja de participar en los foros y enviar e-mails desde su dirección original. Del mismo modo, su cartera digital, estimada en aproximadamente un millón de bitcoins, queda en el limbo y hasta el día de hoy no se han registrado movimientos en ella.

Desde su desaparición muchos nombres han salido a colación como posibles candidatos tras la máscara de Satoshi Nakamoto, pero quizás los más importantes han sido (hasta el momento) Dorian Satoshi Nakamoto, un hombre japonés-americano de California señalado por la revista Newsweek; y el polémico empresario australiano Craig Wright, quien ha afirmado, de hecho, ser Nakamoto, aunque a fin de cuentas no presentó ninguna prueba contundente. En ambos casos se demostró que el seudónimo Satoshi Nakamoto sigue intacto,inlcuso estos escenarios hicierón que Nakamoto volviera a enviar textos cortos desde su viejo correo desmintiendo ser Dorian Nakamoto y Craig Wright, respectivamente.


### 1.3 Una nueva Filosofia


> Bitcoin es descentralizado, son los propios usuarios los que toman las decisiones.


Como señalan Sanchéz y Díaz (2012) en su libro * "BITCOIN: Una moneda criptográfica" *, el Bitcoin es novedoso porque funciona sin una autoridad central que regule la emisión de moneda (generalmente los Bancos Centrales), o acepte o deniegue transacciones, es decir, es descentralizado. Esto es gracias a su arquitectura distribuida, donde son los usuarios del sistema Bitcoin los que implícitamente toman estas decisiones globales “democráticamente”. 

Bajo esta filosofia descentralizada que adopta Bitcoin, son los propios usuarios los que toman las decisiones que normalmente corresponden a una única autoridad central. Esto hace que Bitcoin sea una moneda “democrática”. Como en cualquier democracia, su evolución se adapta a lo que la mayoría de la población quiere. No obstante, en este caso no hay una equivalencia de "un usuario = un voto", ya que el peso o ponderación de cada usuario depende de la potencia de cómputo que éste dedica a la red. Así, la ecuación anterior en Bitcoin, sería más bien $ " x\% \: de \: cómputo = x\% \: de \: votos "$. Por lo tanto, siempre y cuando más de un 50% de la potencia de cómputo de la red sea controlada por usuarios honestos, la red seguirá la evolución que estos decidan. La idea puede contemplarse como una “democracia ponderada” en función de la implicación en el sistema.


Esta nueva filosofia de Bitcoin se puede resumir en los siguientes dos ejemplos:

#### 1. Como recompensa por colaborar con la red, los usuarios reciben bitcoins. 

Hasta aquí, puede parecer que los usuarios podrían engañar al sistema para aumentar su recompensa pero, por construcción del sistema, **la mayoría de los usuarios tendrán que validar** posteriormente esa recompensa. Así, si el usuario la aumentase subrepticiamente, esa acción sería rechazada por el resto.

#### 2. Un usuario A hace un pago con una bitcoin b1 a otro usuario B. 

Para evitar que posteriormente A vuelva a utilizar b1 para pagar a un tercer usuario C, **en Bitcoin, las transacciones se hacen públicas**. Por lo tanto, cuando el resto de la red detecte la segunda transacción, la rechazará, imposibilitando una reutilización de b1 por parte del usuario A.
