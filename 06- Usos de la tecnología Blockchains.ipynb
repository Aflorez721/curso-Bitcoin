## Blockchain: Usos de Bitcoin como tecnología

> La Blockchain como parte del sistema bitcoin, su funcionalidad es básicamente la de proporcionar un registro o libro mayor distribuido en el que se van almacenando las diferentes transacciones realizadas con bitcoins.

Blockchain es la tecnologia que sustenta al Bitcoin. Pero en los ultimos años, esta tecnologia disruptiva a pasado a ser el foco de ínteres para el desarrollo de aplicaciones en diversos campos, entre los que podemos destacar. 

* **Sector Financiero**: 

En el sector financiero, la blockchains presenta aplicaciones que van más allá de Bitcoin, por lo que ha despertado el ínteres de bancos y otras entidades financieras, con lo que se pueden destacar los principales desarrollos:

   + Blockchain presenta la posibilidad de agilizar los pagos y transferencias y el envío de remesas, reduciendo considerablemente los costos que conlleva: por ejemplo, la startup estadounidense *Abra* está desarrollando un sistema digital global de gestión de activos, con funciones de banca minorista como pagos y ahorros y basado en la blockchain de bitcoin que ya permite, entre otras cosas, el envío prácticamente instantáneo de remesas a otros países desde el móvil y por alrededor de un 0,25% del coste actual.
    
   + EL desarrollo de mercados de valores privados basados en tecnologia blockchin, siendo un ejemplo de ello la empresa Nasdaq.
    
   + Mercados de predicción descentralizados: Entre los que se puede destacar Augur, el cuál es un mercado de predicción descentralizado que permite a sus usuarios comprar y vender acciones anticipándose a un suceso en base a la probabilidad de que se produzca uno u otro desenlace.

* **Distribución de Energia:** 

Hoy en día aun es habitual depender de un proveedor central de energía, el cuál  abastece de electricidad a cada casa, empresa o edificio público a cambio de unas tarifas, por lo general abusivas. Sin embargo, a nivel mundial cada vez son más las casas o edificios que generan su propia electricidad con sistemas de energías renovables. Por lo que algunos de estos usuarios de desconectan de la red de suministro electrico, pero muchos tienen a permaner concetados pese a que producen su propia energia. 

> Utilizando una blockchain, las casas y edificios, conectados entre sí a través de una red distribuida, podrían comprar energía a la red o vender sus excedentes dependiendo de sus necesidades en cada momento, sin necesidad de que ningún intermediario lleve el control.

* **Aplicaciones Militares:** 

En cuanto a las aplicaciones militares, la Agencia de Proyectos de Investigación Avanzada de Defensa (DARPA) del Ministerio de Defensa estadounidense y la OTAN han puesto en marcha proyectos relacionados con la aplicación de la tecnología blockchain en sus fuerzas militares.

Un ejemplo de estas aplicaciones militares es la propuesta titulada [“Secure Messaging Platform”](https://sbir.defensebusiness.org/topics/?AspxAutoDetectCookieSupport=1#topic27859) y registrada como parte del Programa SBIR (Small Business Innovation Research Program) el cual tiene como objetivo desarrollar una plataforma de mensajería capaz de transferir mensajes por medio de un protocolo de seguridad descentralizado que sea seguro a través de múltiples canales.

Otro ejemplo lo constituye, la Agencia de información y comunicaciones de la OTAN, quien está evaluando las propuestas enviadas al [2016 Innovation Challenge](https://www.ncia.nato.int/NewsRoom/Pages/160425_Innovation.aspx). La solicitud de propuestas incluía un apartado titulado *“Aplicaciones militares de las blockchains”*.

Además de estas iniciativas, hay otras aplicaciones:

* El bloqueo o desbloqueo automático de armas o vehículos militares dependiendo de quién trate de manejarlos.

* Aplicación de la tecnología de cadenas de bloque a la contratación y financiamiento de proyectos.


* **Internet de las cosas (IoT):** 

El Internet de las cosas es cada vez más popular y a diario surgen nuevos dispositivos conectados, pero la mayor parte de las plataformas de IoT actuales están basadas en modelos centralizados, en los que una empresa o autoridad central controla las conexiones entre los diferentes dispositivos (se ha comprobado ya que este enfoque no resulta práctico en muchos escenarios). Por lo que blockchain se presenta como una excelente opción para la descentralización del Internet de las cosas (IoT) para satisfacer la necesidad de los dispositivos que requieren intercambiar datos entre sí de forma autónoma.

La tecnología blockchain permite el intercambio de datos de forma segura y fiable al tiempo que elabora un registro inmutable de todos los mensajes intercambiados entre los diferentes dispositivos inteligentes conectados. Donde empresas como IBM ya ha creado una plataforma de este tipo en colaboración con Samsung.

La plataforma, denominada ADEPT ([Autonomous Decentralized Peer To Peer Telemetry](https://www-935.ibm.com/services/multimedia/GBE03662USEN.pdf)), genera una red de dispositivos distribuida, utilizando la tecnología blockchain y tres procolos:

   * BitTorrent para la compartición de archivos.
   * Ethereum, para los contratos inteligentes.
   * Y TeleHash para la mensajería P2P.

Una *startup* denominada [Filament](https://filament.com/) ha creado también un software para un IoT descentralizado que utiliza la blockchain de Bitcoin para asignar a cada dispositivo de la red un ID único.

* **Seguros:** 

La combinación de la blockchain con los contratos inteligentes y el IoT podrían revolucionar completamente el sector de los seguros y proporcionar a los usuarios un sistema de gestión de demandas más transparente, responsable e indiscutible.

Bastaría con registrar la condiciones del contrato en un contrato inteligente y almacenarlo en la blockchain, de forma que este se ejecute automáticamente cuando se cumplen unas condiciones determinadas establecidas previamente por ambas partes.

A medida que los hogares, vehículos,entre otros; estén cada vez más llenos de dispositivos y sensores conectados al Internet de las cosas, podrá detectarse automáticamente cualquier incidente, evaluar los daños y realizar el pago correspondiente al afectado de acuerdo con los términos del seguro firmado, incluso antes de que el propio afectado se haya dado cuenta de lo sucedido y sin necesidad de que lo solicite expresamente.

Un excelente ejemplo para el desarrollo de aplicaciones blockchain en el área de seguros viene dado por la empresa [Zircon Tech](http://www.zircon.tech/), la cual está trabajando en una solución que será probada por una compañía de seguros en América Latina a finales de 2017. Esta solución permitirá a la compañía de seguros implementar un programa UBI personalizado sin involucrar a múltiples proveedores. Incluye la provisión de dispositivos y / o aplicaciones para la recolección de datos, almacenamiento y procesamiento de datos, y aplicaciones (web y móvil) para la entrada y visualización de datos. Nuestros algoritmos proporcionarán un sistema de puntuación adaptado a las necesidades y la realidad de nuestros clientes. También incluirá, de acuerdo con las tecnologías más avanzadas, un **desarrollo de Blockchain con IBM Hyperledger** para garantizar la transparencia y la seguridad al administrar la información del cliente final y las múltiples partes interesadas.

* **Propiedad Intelectual:** 

Otra aplicación interesante del blockchain es para compartir y gestionar el uso de propiedad intelectual.Por ejemplo, [ascribe](https://www.ascribe.io) es un servicio que utiliza blockchain para permitir a artistas, galerías y coleccionistas registrar, transferir y archivar arte digital. Los artistas registran sus obras en un registro de propiedad intelectual distribuido que es públicamente accesible. Las obras son almacenadas en la nube y acreditadas al artista. Cada versión de la obra recibe un identificador criptográfico único que es inseparable de la versión original.

* **Cadenas de suministros**

Para un producto compuesto, tiende a ser habitual que parte de un producto proceda de lugares o compañías diferentes. Conformandose de esta forma una cadena de suministros hasta llegar a la compañía que ensambla o elabora y comercializa el producto final.

En ocasiones la cadena de suministros es tan larga que resulta difícil hacer un seguimiento completo de todo el proceso ni de la calidad que conlleve.

Compañías como [Provenance.Org](https://www.provenance.org/), [SkuChain](http://www.skuchain.com/) o [Everledger](https://www.everledger.io/) están utilizando ya la tecnología blockchain para hacer este tipo de seguimientos y garantizar la procedencia de distintos productos: desde ingredientes alimentarios o productos de agricultura, a diamantes, obras de arte y, prácticamente, cualquier otra cosa que lo requiera.

De este modo, la blockchain ofrece una solución a las falsificaciones y puede facilitar la trazabilidad de los productos de cara a la obtención de una certificación (por ejemplo, de producto ecológico).

* **Gestión de identidades:** 

La tecnología blockchain, a travás de sus algoritmos criptograficos, permite a los usuarios crear su propia identidad digital a prueba de manipulación. Y para varios expertos futuristas, esta especie de ID basado en blockchain reemplazará pronto a los nombres de usuario y contraseñas en línea. Algunos ejemplos a destacar son:

   + Empresas como Onename, Keybase o ShoCard, ofrecen servicios para acceder a aplicaciones y sitios web, firmar documentos digitales, entre otros, basados en nuestra identidad blockchain.
    
   + La iniciativa *[Bitnation Blockchain Emergency ID](https://refugees.bitnation.co/faq/)*, que básicamente proporciona un ID digital de emergencia a los inmigrantes y refugiados cuyos pasaportes han sido confiscados o extraviados en su camino. Principalmente se basa en una red de confianza en la que los miembros de cada familia verifican las identidades de los demás miembros.
    
* **Seguridad Automatizada:**

La combinación de las identidades digitales basadas en la blockchain con los contratos inteligentes y las cerraduras electrónicas del Internet de las cosas, permitirá también crear sistemas de seguridad automatizados que garanticen o impidan el acceso a algo de personas concretas de forma completamente automática.

Por ejemplo, se puede almacenar un código en la blockchain para que una puerta de un automovil, de una casa o de una sala dentro de una compañía permita o deniegue el acceso a ella a la persona que se aproxime dependiendo de quién sea la persona que está tratando de acceder.

Esto garantiza completamente la seguridad, dado que nadie tiene que custodiar las llaves ni hay ningún tercero implicado en el proceso que pueda corromperse.



## El proyecto Ethereum 


El proyecto [Ethereum](https://www.ethereum.org/), es una plataforma que permite programar y procesar contratos inteligentes, se puede ver como **una implementación mejorada de la Cadena de Bloques de Bitcoin** que ofrece un lenguaje
de especificación de contratos más robusto que Bitcoin. Ethereum cuenta con su propia criptomoneda Ether (ETH), y junto al blockchain ofrece una máquina virtual distribuida donde se ejecuta el código de los contratos. El consumo de la capacidad de cómputo de la máquina virtual se mide en *“Gas”* que es un token interno de la plataforma, es definido por el programador del contrato y debe ser pagado en ETH por el solicitante de la transacción, tiene como fin evitar
ejecutar contratos maliciosos o con errores por largos períodos de tiempo.


Es importante comprender que Ethereum es fundamentalmente un protocolo, una especificación de un conjunto de reglas y estándares especialmente diseñado para transferencias de contratos inteligentes, y este protocolo puede implementarse con diversas tecnologías de la misma manera que hay distintos clientes y servidores web aunque Internet sea una sola. De igual manera, en una red local podría ejecutarse una plataforma Ethereum aislada, o podrían usarse los componentes de Ethereum para poner en marcha una plataforma alternativa.

> Los **Contratos Inteligentes** son protocolos que facilitan, verifican u obligan la negociación y/o cumplimiento de un contrato en código cuyos resultados eventualmente se registran en una Cadena de Bloques, se suelen programar emulando las cláusulas de un contrato tradicional.

> cualquier persona puede crear un contrato en Ethereum sin necesidad de un servidor centralizado.


### Aplicaciones Descentralizadas


Del inglés DAOs (Decentralized Autonomous Organizations) son el verdadero valor de la plataforma Ethereum. La implantación de DAOs es posible gracias a los tokens personalizados gestionados de forma segura, distribuida y transparente en la cadena de bloques que sirven de base al desarrollo de aplicaciones descentralizadas.
Gartner en su reporte *“Top 10 Strategic Technology Trends for 2017”* (Cearley, Walker, y  Burke, 2016) además de señalar la inteligencia artificial y el aprendizaje automático (machine learning) como la Tendencia No 1, las aplicaciones inteligentes como la tendencia No 2, los objetos inteligentes como la Tendencia No 3, y los “gemelos digitales” como a Tendencia No 5.

Indican en este informe que los elementos claves de estas tecnologías son:

   * Que es un mecanismo de añadir confianza en un entorno de desconfianza,
   * Que se trata de gestionar un registro irrevocable de datos y eventos significativos, que pueden ser transacciones monetarias, registros de propiedad, y otros activos de valor.
   * No consiste solamente en un registro pasivo, sino que es posible añadir eventos automáticos.
