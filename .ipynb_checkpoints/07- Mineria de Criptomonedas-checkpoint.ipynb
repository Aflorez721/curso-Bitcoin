# Mineria con Bitcoin

De  manera  similar  a  como  los  mineros  extraen  oro de  las  minas,  los  bitcoins  son  generados  por  personas,  llamados  mineros, que ejecutan  en  sus  máquinas  un  software  encargado  de  resolver  complejos  cálculos matemáticos   basados  en   criptografía   con   el   fin   de   procesar   y   validar   las transacciones. 

A medida que estos cálculos matemáticos van siendo resueltos, se da  lugar  a  la  creación  de  un  nuevo  bloque  de   bitcoins,  resultado   que  es comunicado  y  registrado  inmediatamente  en  toda  la  red, lo  que  implica  que todos los mineros están compitiendo entre sí a fin de ser los primeros en  resolver estos  cálculos  matemáticos  y  hacerse  con  la  propiedad  de  un  nuevo  bloque de bitcoins.

> **Los mineros** son todos los nodos de la red P2P

![alt text](https://github.com/almus-bot/curso-Bitcoin/tree/master/images/esquema_mineria.png "Esquema de mineria")

También se pueden ver a los mineros como  usuarios especiales que dedican potencia de cómputo a validar nuevas transacciones, creando lo que se conoce como bloques de transacciones. Los cálculos que tienen que realizar son muy costosos por lo que se ven recompensados por ellos.

> En sus inicios, dada la dificultad para obtener bitcoins por medio de minería era tan baja que muchos acumulan cientos por día empleando sus CPU para minar sin darles mayor importancia. Inlcusive abundan los discos duros con abultadas billeteras que acaban descartados o formateados. Miles de bitcoins se pierden en esta etapa, pero casi nadie se lamenta por ello.

Debido a los incrementos en la dificultad para obtener bitcoins de esta manera – *relacionados con el creciente poder computacional que aporta el total de los mineros *– hoy en día es necesario: 

* Contar con potentes tarjetas de video o hardware especializado como los Circuitos Integrados de Aplicación Específica (ASIC por sus siglas en ingles).
* Participar en un pool minero, para *“extraer”* una cantidad apreciable de bitcoins. 

> Hace mucho tiempo que la minería dejó de estar al alcance del usuario común de una PC.

Pero a pesar de las dificultades, muchas personas aun quieren probar la minería de monedas criptográficas, ya sea como curiosidad, o para ver si consigues algún dinero, y se puede afirmar que el proceso no es excesivamente complicado.

> Cualquiera puede convertirse en minero de Bitcoin utilizando un software y hardware especializado.

1. Hardware adecuado. 

El primer paso primordial es contar con el hardware adecuado. Como ya se ha mencionado, en los inicios de la mineria de bitcoin se podria emplear un ordenador estándar (CPU)  o con una tarjeta gráfica (también conocido como minería GPU). Hoy, sin embargo, el nicho de la minería se ha vuelto tan competitivo que **tendrás que utilizar mineros ASIC**, equipos especiales fabricados exclusivamente para la minería Bitcoin. 

> Aunque cuentes con una minería GPU muy potente, ya no puede competir con los chips ASIC incorporados en equipos especiales para minar.

Este tipo de hardware se evalúa según el número de Gigahash por segundo que procesa. Por lo que es fundamental minar los bitcoins con el mejor hardware de minería de bitcoin creado específicamente para ese propósito.

2. Elegir un Pool 


Los grupos de minería de Bitcoin (Pool) son grupos de mineros de Bitcoin que trabajan juntos para resolver un bloque y compartir sus recompensas. **Sin un grupo de minería de Bitcoin, puede que mines bitcoins durante más de un año y nunca ganar ningún bitcoin**. Es mucho más optimo compartir el trabajo y dividir la recompensa con un grupo mucho más grande de mineros de Bitcoin. 

> Actualmente es poco probable conseguir un Bitcoin por ti mismo, por lo que se usan los llamados "mining pools".

Para ver los pools más populares se puede ingresar en <https://blockchain.info/pools>

3. El Software para mineria


Una vez que se obtiene el hardware especializado para minar, se necesita de un software especializado que se conecte con la red bitcoin, más especificamente a la red P2P. La elección del software dependera de qué equipo de minería tengas, tendrás que encontrar el software adecuado. Para lo cual se presentan varias opciones:
