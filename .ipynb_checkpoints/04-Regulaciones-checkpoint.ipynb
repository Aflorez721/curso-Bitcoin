## 8. Regulaciones de bitcoin en el mundo


A nivel de regulaciones se debe considerar que **el bitcoin no es una moneda oficial**. 

En la mayoría de paises, los ciudadanos tendrán que pagar renta, nóminas y los impuestos sobre ganancias en todo lo que tenga valor, incluyendo bitcoins. Es responsabilidad de los ciudadanos asegurarse de aplicar los impuestos y otras regulaciones publicadas por su gobierno o municipio. 

Para el año 2016, a nivel de regulaciones para el uso de Bitcoin, la mayoría de los países del mundo tienen legislación neutra o abiertamente favorable sobre bitcoin. La tendencia tiende a verse positiva, las leyes son cada vez más pro-bitcoin. Las autoridades financiaras de paises como Reino Unido, Holanda, Japón, Nueva Zelanda, India e Israel han hecho afirmaciones muy favorables a bitcoin y las criptodivisas. El Parlamento Europeo aprobó en mayo de 2016 una resolución muy equilibrada sobre monedas virtuales reconociendo tanto sus grandes ventajas como sus posibles riesgos (Grupo BTC,2016).

Por otro lado, países como Bangladesh, Ecuador y Bolivia, prohíben el uso de bitcoin y el resto de monedas virtuales, mientras que países como Rusia mantienen posturas poco claras. Obsérvese en todo caso que ninguna de estas limitaciones es intrínseca a bitcoin, sino debida principalmente a su estadio inicial. Un crecimiento gradual en el mercado del bitcoin necesariamente provocará un decrecimiento de estos inconvenientes.

* **Normativa en Japón**

Japón es quiza la mayor referencia en proyectos y avances regulatorios de blockchain y, especialmente, de Bitcoin. 

El 07 de febrero de 2017 la Agencia de Servicios Financieros de Japón, también conocida como FSA (Financial Services Agency), dio a conocer un [comunicado público](http://www.fsa.go.jp/news/28/ginkou/20170324-1.html) sobre la aplicación de la ley de modificación de las ordenanzas de préstamos de dinero y de la ley de la banca por decreto ministerial sugerido por los abogados de la firma *Anderson Mori & Tomotsune*.

Dicha ley solo **contempla a Bitcoin como método de pago, más no como una moneda propiamente dicha.**

> Desde el primero de abril de 2017 en Japón se acepta el Bitcoin como medio de pago.

La adopción de Bitcoin como medio de pago se somete a un grupo de regulaciones acordadas por la FSA, las cuales se extenderán a las casas de cambio y cualquier tipo de servicio que maneje dinero virtual de las personas.

Dichas regulaciones abarcan:

* Requisitos administrativos y legales para el préstamo de servicio por parte de cualquier negocio en Japón

* Normativas de prevención de transferencias de fondos de actividades ilícitas.

  + las normas de identidad del cliente (KYC) 
  + Normas de anti-lavado de dinero (AML).


* **Normativa en Rusia**
 
En Rusia, para este 2017 se continua con las posiciones encontradas entre autoridades bancarias y gubernamentales sobre cómo proceder con la adopción de criptoactivos. En septiembre de 2017 la directora del Banco Central de Rusia ha ofrecido declaraciones de desconfianza e incluso rechazo hacia la adopción de criptomonedas, mientras que por su parte el presidente Putin y el ministro de finanzas han estado evaluando acciones para permitir el uso de bitcoins y otras monedas digitales en el Mercado de Acciones de Moscú (MOEX).


En el II Foro Financiero de Moscú, el ministro de Finanzas de Rusia,Anton Siluanov, expresó  que el gobierno no tiene intenciones de ilegalizar las criptomonedas: “El Estado entiende que las criptomonedas son un hecho. No tiene sentido prohibirlas, hay que regularlas”.


* **Normativa en Estados Unidos**

Estados Unidos es uno de los paises de mayor poder económico en el mundo. Y el Bitcoin ha recibido un amplio apoyo de empresarios, desarrolladores y autoridades estadounidenses, conformando un gran aporte al desarrollo del ecosistema blockchain, por lo que la adopción del bitcoin ha tomado mayor fuerza cada día.

Este aumento en el uso de Bitcoin y otras criptomonedas también se traduce en una mayor preocupación por parte de las autoridades estadounidenses sobre las faltas de regulaciones que para antes del año 2012  eran inexistentes, aunado además a que las *coin* representan un cambio conceptual del sistema tradicional financiero que domina el mercado norteamericano.


El estado de Nueva York en 2015 fue uno de los primeros en adoptar marcos regulatorios sobre el bitcoin, con el llamado [Bitlicense](http://www.dfs.ny.gov/legal/regulations/adoptions/dfsp200t.pdf), que  es el término común usado para una licencia comercial de actividades de moneda virtual, emitida por el Departamento de Servicios Financieros del Estado de Nueva York (NYSDFS) bajo regulaciones diseñadas para compañías. Medida que afecto a compañías de Bitcoin a nivel mundial, quienes cortaran sus servicios con sus clientes residentes en New York por considerar qu BitLicense pondría la privacidad de nuestros usuarios en peligro y comprometería la privacidad de sus empleados (por tener que enviar sus huellas dactilares al NYDFS).

> La Bitlicense busca ofrecer protección al consumidor, evitar lavado de dinero y establecer unas reglas de ciberseguridad adaptadas a las compañías que trabajan con monedas digitales.


Ya en 2016 la elección del presidente Donald Trump frente a Hillary Clinton, ocasionó un gran impacto en el sector de las criptomonedas, impacto que inicia incluso desde la campaña presidencial donde ambos candidatos contemplaban propuestas para la adopción de la tecnologia blockchain. Siendo Trump quien se perfiló como el mejor candidato que pudiese impulsar más firmemente la adopción de las tecnologías distribuidas y de las criptomonedas debido a iniciativas y personalidades relacionadas a Bitcoin que han ayudado y caracterizado su oferta presidencial.

> Con Trump en la presidencia se empezó a dibujar un panorama distinto en cuanto a blockchain y criptomonedas en los Estados Unidos, más aun con la presencia de tres miembros del gobierno abiertamente pro-bitcoin.

 
* **Normativa en Mexico**

Las autoridades financieras de Mexico avanzan desde marzo de 2017 en el desarrollo de la Ley de Tecnología Financiera (Fintech). Este proyecto de ley se concibe para regular la tecnología financiera en Mexico, donde en un principio se buscaba regular a las entidades que compran y venden activos virtuales como el bitcoin.

> El anteproyecto de ‘Ley Fintech’ se concentra principalmente se enfocará principalmente a regular lo que se conoce como Robo Advisory o asesorías digitales a través de plataformas tecnológicas y que utilizan algoritmos para comunicarse entre clientes y entidades financieras.

Por otro lado, instituciones como la Reserva Federal de los Estados Unidos, la Comisión de la Bolsa y Valores y el FBI, todavía no están del todo convencidas sobre el funcionamiento masivo de las monedas digitales descentralizadas y dedican sus esfuerzos a la investigación de la tecnologia blockchain. 

Además el ecosistema de Bitcoin debe lidiar con los desacuerdos estatales, donde las regulaciones regionales también hacen de lo suyo en el mercado de las monedas criptográficas; algunos departamentos estan solicitando a las empresas la adquisición de una licencia regional que le permita realizar transacciones financieras, mientras que otros están más abiertas a la posibilidad de adoptar blockchain y las criptomonedas. 

* **Normativa en Europa**

No existe en España (ni en muchos otros países europeos) regulación específica para las criptomonedas. Este hecho no debe transmitir la falsa seguridad del vacío legal. Si proveedores, clientes, inversores o hacienda se ven perjudicados por actividad bitcoin, el poder judicial encontrará formas de aplicar las leyes existentes. No se debe olvidar que desde principios de siglo se ha desarrollado una especial sensibilidad contra el blanqueo de capitales. Por tanto en la medida de lo posible, las autoridades querrán trazar todo movimiento de dinero, independientemente de su naturaleza.

Sin embargo, el Banco Central Europeo (BCE), ha tomado una posición de neutralidad y cautela ante las herramientas financieras. Donde el presidente del BCE, Mario Draghi, a indicado que **la autoridad bancaria europea no tiene facultades para la regulación o prohibición de criptoactivos.**

* **Normativa en China**

China es quiza el país que ha venido desarrollando una de las normativas más rigidas al mercado de criptomonedas. Sin embargo, también es, junto a Japón, uno de los principales tenedores de Bitcoins del mundo.

> China supone el $23\%$ de la negociación de bitcoins en el mundo, y acoge los mayores centros de minería de bitcoins. 
<https://cincodias.elpais.com/cincodias/2017/09/15/mercados/1505463032_514339.html>

Para la primera semana de septiembre de 2017, el Banco Central de China (BCC) ha anunciado la prohibición inmediata de las inversiones basadas en ICO (Ofertas Iniciales de Moneda), que según este organismo *"han perturbado seriamente el orden económico y financiero"*, al  considerar ilegal el **crowdfunding** a través de bitcoin y ha exigido el cese de toda recaudación de fondos relacionadas con la criptomoneda. 

El BCC también le ordeno a  los mercados de cambio de este tipo de criptomonedas, que no permitan más el registro de nuevos usuarios a partir del viernes 08 de septiembre, según un comunicado del Gobierno firmado por el grupo de la ciudad de Pekín a cargo de supervisar riesgos financieros en Internet, que circuló en Internet. 

Por otro lado, ya que su idea es seguir innovando en el uso de las monedas digitales.


* **Normativa en Colombia**

En Colombia la normativa regulatoria de las criptomonedas tiende a ser difuso. La Superintendencia Financiera de Colombia ha expresado en varios comunicados, que el peso colombiano es la única divisa de curso legal, por lo que las criptomonedas no son reconocidas como divisas en el país. 

Sin embargo, en Colombia aun se permite, hasta el momento, las operacones de negocios con criptomonedas, como SurBTC y se han realizado eventos que promueven el uso de bitcoin, tal como lo es el **Bitcoin Airdrop**, así como la instalación de un cajero de criptomonedas instalado por Compudemano en la ciudad de Pereira.

También cabe destacar que el 29 de agosto de 2017, el Banco de la República de Colombia anunció su asociación con el **consorcio blockchain R3** para explorar la aplicación de su plataforma Corda en el intercambio de valores del banco, por lo que, si bien aún el Estado colombiano se mantiene reticente a las criptomonedas, no es el mismo caso con la tecnología que le subyace.


* **Normativa en Venezuela**

En Venezuela **no existe un marco legal sobre Bitcoin ni la tecnología blockchain**, sin embargo el ministerio de Relaciones Interiores, Justicia y Paz (MREJP) ha realizado investigaciones para *“debilitar operaciones en criptomonedas”* principalmente orientado a detectar mineros en el país.

Durante los años 2016 y 2017 ya se han presentado casos de arresto de personas bajo la sospecha de delitos electrónicos, legitimación de capitales, asociación para delinquir y contrabando, donde según informa el mismo SEBIN, su actividad era minar bitcoins.

Quizas el caso más sonado ha sido el desmantelamiento de un galpón en Carabobo, dedicado a la minería de critopmonedas con cerca de 11.000 equipos y en el que se detuvieron a dos jóvenes pertenecientes a una red que opera desde Polonia y se dedica al fraude cambiario, lavado de dinero y delitos informáticos.
